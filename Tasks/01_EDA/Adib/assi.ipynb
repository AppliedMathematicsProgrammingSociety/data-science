```python
import matplotlib.pyplot as mp
import seaborn as sns
mp.figure(figsize=(15,5))
sns.scatterplot(x="votes", y="approx_cost(for two people)",data=data_collected,hue="name")
```

```python
for col in ["online_order", "book_table","approx_cost(for two people)","listed_in(type)"]:
    sns.countplot(data_collected, x=col,edgecolor='black',palette="Set2")
    mp.title("Distribution of " + col)
    mp.xlabel(col)
    mp.xticks(rotation=90)
    mp.show()
```
