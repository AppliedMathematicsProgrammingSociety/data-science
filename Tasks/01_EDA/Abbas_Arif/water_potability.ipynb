```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

## Description of the dataset

```python
df = pd.read_csv('water_potability.csv')
df.info()
```

## Handling missing values

```python
missing = ["Sulfate", "ph", "Trihalomethanes"]
for cols in missing:
    m = df[cols].mean()
    df[cols].fillna(m, inplace = True)
df.info()

        
```

## Numerical Analysis

```python
numvar = ["ph", "Hardness", "Solids", "Chloramines", "Sulfate", "Conductivity", "Organic_carbon", "Trihalomethanes", "Turbidity"]
catvar = ["Potability"]
for vars in numvar:
    print(f"Information on {vars}")
    print(df[vars].describe())
```

### Visualization

```python

fig, axes = plt.subplots(3, 3, figsize=(12, 15))
axes = axes.ravel()

for i, vars in enumerate(numvar):
    axes[i].hist(df[vars], bins=50)
    axes[i].set_xlabel(vars)
  
    axes[i].set_title(f"Information on {vars}")
plt.tight_layout()
plt.show()


```

```python
for vars in numvar:
    plt.hist(df[vars],bins=10,edgecolor='black',color='blue')
    plt.title(f"{vars} vs values histogram")
    plt.xlabel(vars)
    plt.show()
```

## Categorical Analysis


```python
numvar = ["ph", "Hardness", "Solids", "Chloramines", "Sulfate", "Conductivity", "Organic_carbon", "Trihalomethanes", "Turbidity"]
catvar = ["Potability"]



for i in catvar:
    print(df[i].value_counts())
    print(df[i].value_counts(normalize=True) * 100)
    print(df[i].unique())

for vars in numvar:
    print(df.groupby("Potability")[vars].mean())
    print(df.groupby("Potability")[vars].median())


    
```
